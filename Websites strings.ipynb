```python
from bs4 import BeautifulSoup
import urllib
import re
```

```python
html_doc1 = urllib.urlopen("https://en.wikipedia.org/wiki/Lego").read()
soup1 = BeautifulSoup(html_doc1, 'html.parser')
```

```python
html_doc2 = urllib.urlopen("https://en.wikipedia.org/wiki/Oil").read()
soup2 = BeautifulSoup(html_doc2, 'html.parser')
```

```python
for string in soup1.stripped_strings:print(repr(string))
    
```

```python
#list_topic = list()
file = open('ParsedString3.txt','wb' )
for strings2 in soup1.stripped_strings:
    strings2 =  repr(strings2)
    strings2 = str(strings2)
    file.write(strings2)
file.flush()
file.close()
```

```python
file = open('ParsedString4.txt','wb' )
for link in soup2.stripped_strings:
    ## First convert to string
    link =  repr(link)
    link = str(link)
    file.write(link)
file.flush()
file.close()
```
